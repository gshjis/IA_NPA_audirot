```python
# ============================================================================
# СЕМАНТИЧЕСКИЙ ТЕСТ КОНСТИТУЦИИ
# ============================================================================

print("=" * 80)
print("🔍 ТЕСТИРОВАНИЕ СЕМАНТИЧЕСКОГО ПОИСКА".center(80))
print("=" * 80)

for idx, query in enumerate(test_queries, 1):
    
    print(f"\n{'-' * 80}")
    print(f"📌 ЗАПРОС №{idx}: {query}")
    print(f"{'-' * 80}")
    
    # Поиск топ-3 статей
    results = tagger.find_articles_by_new_sentence(query, k=3)
    
    if not results:
        print("❌ НИЧЕГО НЕ НАЙДЕНО")
        continue
    
    for rank, res in enumerate(results, 1):
        # Определяем иконку по score
        if res['score'] >= 0.6:
            icon = "🟢🔝"
        elif res['score'] >= 0.5:
            icon = "🟡"
        elif res['score'] >= 0.4:
            icon = "🔸"
        else:
            icon = "⚪"
        
        print(f"\n{icon} [{rank}] РЕЛЕВАНТНОСТЬ: {res['score']:.3f}")
        print(f"   {res['text'][:300]}...")
    
    print(f"\n{'-' * 80}")

print("\n" + "=" * 80)
print("🏁 ТЕСТИРОВАНИЕ ЗАВЕРШЕНО".center(80))
print("=" * 80)
```
