```python
import bm25s
import Stemmer
from loader import *
```

```python
stemmer = Stemmer.Stemmer("russian")

raw_D = load_json("/home/gshjis/Python_projects/IA_NPA_auditor/data/raw/constitution.json")
D_raw = []
for r in raw_D:
    for art in r["articles"]:
        D_raw.append(art["text"])  # сырой текст, без предобработки

corpus = D_raw  # используем сырые тексты
```

```python
# BM25S сделает все сам
corpus_tokens = bm25s.tokenize(
    corpus,
    stopwords="ru",
    stemmer=Stemmer.Stemmer("russian")
)

# 3. Создание индекса
retriever = bm25s.BM25()
retriever.index(corpus_tokens)
```
